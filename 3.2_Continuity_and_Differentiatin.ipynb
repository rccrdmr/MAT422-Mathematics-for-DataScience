# 3.2 Continuity and Differentiaton

### 3.2.1 Limits and Continuity

Limits play a crucial role in calculus and mathematical analysis, helping to define concepts such as continuity, derivatives, and integrals. The Euclidean norm for a vector $x \in \mathbb{R}^d$ is defined as:

$$
\|x\| = \sqrt{\sum_{i=1}^d x_i^2}
$$

An open $r$-ball around a point $x \in \mathbb{R}^d$ is the set of points within a Euclidean distance $r$ of $x$:

$$
B_r(x) = \{ y \in \mathbb{R}^d : \|y - x\| < r \}
$$

Screenshot 2024-10-13 at 2.06.34 PM.png

----

**Definition (Limits of a Function)**

Let $f: D \to \mathbb{R}$ be a real-valued function on $D \subseteq \mathbb{R}^d$. $f$ has a limit $L$ as $x$ approaches $a$ if, for any $\epsilon > 0$, there exists a $\delta > 0$ such that $|f(x) - L| < \epsilon$ for all $x \in D \cap B_\delta(a) \setminus \{a\}$. This is written as:

$$
\lim_{x \to a} f(x) = L
$$

----

**Definition (Continuous Function)**

A function $f: D \to \mathbb{R}$ is continuous at $a \in D$ if:

$$
\lim_{x \to a} f(x) = f(a)
$$

A continuous function has no abrupt changes in value, and can be visualized with a graph that doesn't break.
Screenshot 2024-10-13 at 2.02.08 PM.png

----

**Lemma (Composition of Continuous Functions)**

Let $f: D_1 \to \mathbb{R}^m$, where $D_1 \subseteq \mathbb{R}^d$, and $g: D_2 \to \mathbb{R}^p$, where $D_2 \subseteq \mathbb{R}^m$. If $f$ is continuous at $x_0$ and $g$ is continuous at $f(x_0)$, then $g \circ f$ is continuous at $x_0$.

----

**Definition (Extremum)**

Suppose $f: D \to \mathbb{R}$ is defined on $D \subseteq \mathbb{R}^d$. $f$ attains a maximum value $M$ at $z^*$ if $f(z^*) = M$ and $M \geq f(x)$ for all $x \in D$. Similarly, $f$ attains a minimum value $m$ at $z^*$ if $f(z^*) = m$ and $m \leq f(x)$ for all $x \in D$.

----

**Theorem (Extreme Value)**

Let $f: D \to \mathbb{R}$ be a real-valued, continuous function on a nonempty, closed, bounded set $D \subseteq \mathbb{R}^d$. Then $f$ attains both a maximum and a minimum on $D$.

```python
import sympy as sym

x, y = sym.symbols('x y')

# Calculate the limit of sin(x) / x as x approaches 0
f1 = sym.sin(x) / x
limit1 = sym.limit(f1, x, 0)
print("Limit of sin(x)/x as x approaches 0:", limit1)

# Calculate the limit of ((n+3)/(n+2))^n as n approaches infinity
n = sym.symbols('n')
f2 = ((n + 3) / (n + 2)) ** n
limit2 = sym.limit(f2, n, sym.oo)
print("Limit of ((n+3)/(n+2))^n as n approaches infinity:", limit2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function with a discontinuity
def f1(x):
    return np.where(x != 0, np.sin(x) / x, 1)  # Handling discontinuity at x = 0

x_vals = np.linspace(-10, 10, 400)
y_vals = f1(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r"$\frac{\sin(x)}{x}$", color='blue')
plt.axvline(0, color='red', linestyle='--')
plt.ylim(-0.5, 1.5)
plt.title("Limit and Continuity of $\sin(x)/x$ as $x$ approaches 0")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

### 3.2.2 Derivatives

#### 3.2.2.1 Single-Variable Case

The derivative of a function measures the sensitivity to change of the function value with respect to a change in the input value.

----

**Definition (Derivative)**

For $f: D \to \mathbb{R}$ where $D \subseteq \mathbb{R}$, the derivative of $f$ at $x_0 \in D$ is:

$$
f'(x_0) = \lim_{h \to 0} \frac{f(x_0 + h) - f(x_0)}{h}
$$

If the limit exists, $f$ is differentiable at $x_0$.

----

**Proposition**

If $f$ and $g$ have derivatives at $x$, and $\alpha, \beta$ are constants, then:

$$
[\alpha f(x) + \beta g(x)]' = \alpha f'(x) + \beta g'(x)
$$

----

**Lemma**

Let $f: D \to \mathbb{R}$ with $D \subseteq \mathbb{R}$ and $x_0 \in D$ where $f'(x_0)$ exists. If $f'(x_0) > 0$, then there is an open ball $B_\delta(x_0) \subseteq D$ around $x_0$ such that:

- $f(x) > f(x_0)$ if $x > x_0$
- $f(x) < f(x_0)$ if $x < x_0$

If $f'(x_0) < 0$, the opposite holds.

----

**Theorem (Rolle)**

Let $f: [a, b] \to \mathbb{R}$ be a continuous function, and assume its derivative exists on $(a, b)$. If $f(a) = f(b)$, then there exists $c \in (a, b)$ such that:

$$
f'(c) = 0
$$

Screenshot 2024-10-13 at 3.05.10 PM.png

----

**Theorem (Mean Value)**

Let $f: [a, b] \to \mathbb{R}$ be a continuous function, and assume its derivative exists on $(a, b)$. Then there exists $c \in (a, b)$ such that:

$$
f(b) - f(a) = (b - a) f'(c)
$$

Or equivalently:

$$
\frac{f(b) - f(a)}{b - a} = f'(c)
$$

```python
x = sym.symbols('x')

# Calculate the first, second, and third derivatives of sin(2x)
f3 = sym.sin(2 * x)
deriv1 = sym.diff(f3, x)
deriv2 = sym.diff(f3, x, 2)
deriv3 = sym.diff(f3, x, 3)

print("First derivative of sin(2x):", deriv1)
print("Second derivative of sin(2x):", deriv2)
print("Third derivative of sin(2x):", deriv3)
```

```python
x_vals = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y_vals = np.sin(2 * x_vals)
y_prime = 2 * np.cos(2 * x_vals)
y_double_prime = -4 * np.sin(2 * x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r"$\sin(2x)$", color='blue')
plt.plot(x_vals, y_prime, label=r"$2\cos(2x)$ (1st derivative)", color='green')
plt.plot(x_vals, y_double_prime, label=r"$-4\sin(2x)$ (2nd derivative)", color='orange')
plt.title("Function and its Derivatives")
plt.xlabel("x")
plt.ylabel("f(x) and derivatives")
plt.legend()
plt.grid()
plt.show()
```

#### 3.2.2.2 General Case

For functions of multiple variables, partial derivatives measure the rate of change of the function with respect to each variable independently.

----

**Definition (Partial Derivative)**

For $f: D \to \mathbb{R}$ where $D \subseteq \mathbb{R}^d$ and $x_0 \in D$, the partial derivative of $f$ at $x_0$ with respect to $x_i$ is:

$$
\frac{\partial f(x_0)}{\partial x_i} = \lim_{h \to 0} \frac{f(x_0 + h e_i) - f(x_0)}{h}
$$

If $\frac{\partial f(x_0)}{\partial x_i}$ exists and is continuous in an open ball around $x_0$ for all $i$, then $f$ is continuously differentiable at $x_0$.

----

**Definition (Jacobian)**

Let $f = (f_1, \dots, f_m): D \to \mathbb{R}^m$ where $D \subseteq \mathbb{R}^d$ and $x_0 \in D$. The Jacobian of $f$ at $x_0$ is the $d \times m$ matrix:

$$
J_f(x_0) = \begin{pmatrix}
\frac{\partial f_1(x_0)}{\partial x_1} & \dots & \frac{\partial f_1(x_0)}{\partial x_d} \\
\vdots & \ddots & \vdots \\
\frac{\partial f_m(x_0)}{\partial x_1} & \dots & \frac{\partial f_m(x_0)}{\partial x_d}
\end{pmatrix}
$$

----

**Example**

Consider the affine function $f(x) = q^T x + r$ where $x = (x_1, \dots, x_d)^T$ and $q \in \mathbb{R}^d$. The partial derivatives of the linear term are given by:

$$
\frac{\partial}{\partial x_i} [q^T x] = q_i
$$

Thus, the gradient of $f$ is:

$$
\nabla f(x) = q
$$


----

**Theorem (Chain Rule)**

Let $f: D_1 \to \mathbb{R}^m$, where $D_1 \subseteq \mathbb{R}^d$, and let $g: D_2 \to \mathbb{R}^p$, where $D_2 \subseteq \mathbb{R}^m$. Assume $f$ is continuously differentiable at $x_0$, and $g$ is continuously differentiable at $f(x_0)$. Then:

$$
J_{g \circ f}(x_0) = J_g(f(x_0)) J_f(x_0)
$$

This expresses the Jacobian of the composition as the product of the Jacobians.

```python
x, y, z = sym.symbols('x y z')

# Compute partial derivatives of exp(x*y*z)
f4 = sym.exp(x * y * z)
partial_x = sym.diff(f4, x)
partial_y = sym.diff(f4, y)
partial_z = sym.diff(f4, z)

print("Partial derivative of exp(x*y*z) w.r.t x:", partial_x)
print("Partial derivative of exp(x*y*z) w.r.t y:", partial_y)
print("Partial derivative of exp(x*y*z) w.r.t z:", partial_z)

# Calculate Jacobian for a function with variables (rho, phi)
rho, phi = sym.symbols('rho phi')
X = sym.Matrix([rho * sym.cos(phi), rho * sym.sin(phi), rho ** 2])
Y = sym.Matrix([rho, phi])
jacobian = X.jacobian(Y)

print("Jacobian matrix:", jacobian)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Multivariable function
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = X * np.exp(-X**2 - Y**2)

# Plot the 3D surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface of $x e^{-(x^2 + y^2)}$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.show()

# Contour plot and gradient vectors
plt.figure(figsize=(10, 6))
contours = plt.contour(X, Y, Z, levels=15, cmap='viridis')
plt.quiver(X, Y, np.gradient(Z, axis=0), np.gradient(Z, axis=1))
plt.colorbar(contours)
plt.title("Contour Plot with Gradient of $x e^{-(x^2 + y^2)}$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

#### 3.2.2.3 Further Derivatives

The directional derivative measures the rate of change of a multivariable function along a specified direction.

----

**Definition (Directional Derivative)**

For $f: D \to \mathbb{R}$ where $D \subseteq \mathbb{R}^d$, $x_0 \in D$, and a unit vector $v \in \mathbb{R}^d$, the directional derivative of $f$ at $x_0$ in the direction $v$ is:

$$
\frac{\partial f(x_0)}{\partial v} = \lim_{h \to 0} \frac{f(x_0 + h v) - f(x_0)}{h}
$$

----

**Theorem (Directional Derivative from Gradient)**

Let $f: D \to \mathbb{R}$ where $D \subseteq \mathbb{R}^d$, $x_0 \in D$, and $v \in \mathbb{R}^d$ is a unit vector. If $f$ is continuously differentiable at $x_0$, then the directional derivative of $f$ at $x_0$ in the direction $v$ is given by:

$$
\frac{\partial f(x_0)}{\partial v} = \nabla f(x_0)^T v
$$

----

**Definition (Second Partial Derivatives and Hessian)**

For $f: D \to \mathbb{R}$ where $D \subseteq \mathbb{R}^d$, the Hessian matrix at $x_0$ is:

$$
H_f(x_0) = \begin{pmatrix}
\frac{\partial^2 f(x_0)}{\partial x_1^2} & \dots & \frac{\partial^2 f(x_0)}{\partial x_d \partial x_1} \\
\vdots & \ddots & \vdots \\
\frac{\partial^2 f(x_0)}{\partial x_1 \partial x_d} & \dots & \frac{\partial^2 f(x_0)}{\partial x_d^2}
\end{pmatrix}
$$

If $f$ is twice continuously differentiable at $x_0$, the Hessian is symmetric.

----

**Theorem (Symmetry of the Hessian)**

Let $f: D \to \mathbb{R}$ be twice continuously differentiable at $x_0 \in D$. Then:

$$
\frac{\partial^2 f(x_0)}{\partial x_j \partial x_i} = \frac{\partial^2 f(x_0)}{\partial x_i \partial x_j}
$$

----

**Example**

Consider the quadratic function $f(x) = \frac{1}{2} x^T P x + q^T x + r$, where $P$ is a symmetric matrix in $\mathbb{R}^{d \times d}$, $q \in \mathbb{R}^d$, and $r \in \mathbb{R}$. The gradient of $f$ is:

$$
\nabla f(x) = P x + q
$$

The Hessian of $f$ is:

$$
H_f(x) = P
$$

Since $P$ is symmetric, the Hessian is also symmetric. This symmetry verifies the theorem on the symmetry of second partial derivatives.


```python
x, y, z = sym.symbols('x y z')

# Calculate the directional derivative of x^2 - y - z^2
f5 = x ** 2 - y - z ** 2
gradient = [sym.diff(f5, var) for var in (x, y, z)]
direction = sym.Matrix([1, -1, 0])  # example unit direction vector
directional_deriv = sum([g * d for g, d in zip(gradient, direction)])

print("Directional derivative in direction (1, -1, 0):", directional_deriv)

# Compute the Hessian matrix of the function
hessian = sym.hessian(f5, (x, y, z))
print("Hessian matrix:", hessian)
```

```python
# Define a function and plot its gradient field
def f2(x, y):
    return x ** 2 - y ** 2

x_vals = np.linspace(-2, 2, 20)
y_vals = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f2(X, Y)

dx, dy = np.gradient(Z)

plt.figure(figsize=(10, 6))
plt.contour(X, Y, Z, levels=15, cmap='coolwarm')
plt.quiver(X, Y, dx, dy, scale=50)
plt.title("Directional Derivative and Gradient Field of $x^2 - y^2$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

### 3.2.3 Taylor's Theorem

Taylor’s Theorem provides an approximation of a differentiable function around a given point using a polynomial expansion. We focus on the linear approximation with a second-order error term.

----

**Theorem (Taylor)**

If $f: D \to \mathbb{R}$ where $D \subseteq \mathbb{R}$, and $f$ has $m$ continuous derivatives on $[a, b]$, then:

$$
f(b) = f(a) + (b - a) f'(a) + \frac{(b - a)^2}{2!} f''(a) + \dots + \frac{(b - a)^{m-1}}{(m-1)!} f^{(m-1)}(a) + R_m
$$

where the remainder $R_m$ is given by:

$$
R_m = \frac{(b - a)^m}{m!} f^{(m)}(a + \theta(b - a)), \quad \text{for some } \theta \in (0,1)
$$

In particular, for $m = 2$, we have:

$$
f(b) = f(a) + (b - a) f'(a) + \frac{(b - a)^2}{2} f''(\xi), \quad \text{for some } a < \xi < b
$$

----

**Theorem (Multivariate Mean Value)**

Let $f: D \to \mathbb{R}$ where $D \subseteq \mathbb{R}^d$. Let $x_0 \in D$ and $\delta > 0$ such that $B_\delta(x_0) \subseteq D$. If $f$ is continuously differentiable on $B_\delta(x_0)$, then for any $x \in B_\delta(x_0)$:

$$
f(x) = f(x_0) + \nabla f(x_0 + \xi p)^T p
$$

for some $\xi \in (0, 1)$, where $p = x - x_0$.

----

**Theorem (Multivariate Taylor)**

Let $f: D \to \mathbb{R}$ where $D \subseteq \mathbb{R}^d$. If $f$ is three times continuously differentiable on $B_\delta(x_0)$, then for any $x \in B_\delta(x_0)$:

$$
f(x) = f(x_0) + \nabla f(x_0)^T p + \frac{1}{2} p^T H_f(x_0) p + O(\|p\|^3)
$$

where $p = x - x_0$. If $f$ is twice continuously differentiable, then:

$$
f(x) = f(x_0) + \nabla f(x_0)^T p + \frac{1}{2} p^T H_f(x_0 + \xi p) p
$$

for some $\xi \in (0, 1)$.

----

**Example**

Consider the function $f(x_1, x_2) = x_1 x_2 + x_1^2 + e^{x_1} \cos(x_2)$. We apply Taylor’s Theorem at $x_0 = (0, 0)$ and $x = (x_1, x_2)$. The gradient is:

$$
\nabla f(x_1, x_2) = (x_2 + 2x_1 + e^{x_1} \cos(x_2), x_1 - e^{x_1} \sin(x_2))^T
$$

and the Hessian is:

$$
H_f(x_1, x_2) = \begin{pmatrix}
2 + e^{x_1} \cos(x_2) & 1 - e^{x_1} \sin(x_2) \\
1 - e^{x_1} \sin(x_2) & -e^{x_1} \cos(x_2)
\end{pmatrix}
$$

At $x_0 = (0, 0)$, we have $f(0, 0) = 1$ and $\nabla f(0, 0) = (1, 0)^T$. By Taylor’s Theorem, we approximate:

$$
f(x_1, x_2) \approx 1 + x_1 + \frac{1}{2} \left[ 2x_1^2 + 2x_1 x_2 + (x_1^2 - x_2^2) e^{x_1} \cos(x_2) - 2x_1 x_2 e^{x_1} \sin(x_2) \right]
$$


```python
x, y = sym.symbols('x y')

# Calculate first and second-order Taylor expansion of ln(x^2 + y^2 + 1) at (0, 0)
f6 = sym.ln(x ** 2 + y ** 2 + 1)
taylor_first_order = f6.series(x, 0, 2).removeO().subs(y, 0)
taylor_second_order = f6.series(x, 0, 3).removeO().subs(y, 0)

print("First-order Taylor expansion at (0, 0):", taylor_first_order)
print("Second-order Taylor expansion at (0, 0):", taylor_second_order)
```

```python
# Define the function and Taylor expansions
def f3(x):
    return np.exp(x)

# Taylor expansions
def taylor_approximation(x, order):
    return sum([x**n / np.math.factorial(n) for n in range(order + 1)])

x_vals = np.linspace(-2, 2, 400)
y_vals = f3(x_vals)
y_approx_2 = taylor_approximation(x_vals, 2)
y_approx_4 = taylor_approximation(x_vals, 4)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r"$e^x$", color='blue')
plt.plot(x_vals, y_approx_2, '--', label="2nd Order Taylor Approximation", color='green')
plt.plot(x_vals, y_approx_4, ':', label="4th Order Taylor Approximation", color='orange')
plt.title("Taylor Series Approximation of $e^x$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
